EXERCISE 1:

Plot the function f(x) = sin(x)+cos(x) in the [0,2*pi] interval.

```python
# value range x in [0, 2*pi] using numpy library
x = np.linspace(0, 2*np.pi)
# values of function f(x) = sin(x) + cos(x)
y = np.sin(x) + np.cos(x)
# plot using matplotlib
plt.plot(x, y, label='f(x) = sin(x) + cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) = sin(x) + cos(x)')
plt.legend()
plt.grid(True)
plt.show()
```

Use the function to plot a heatmap that depicts a part of the Mandelbrot fractal. The heatmap must be 600 pixels wide and 400 pixels high. The values of the real part of the complex numbers must range between ‐2 and 1, while the values of the imaginary part must range between ‐1 and 1.


```python
MAX_ITERATIONS = 80

def mandelbrot(a, b):
    c = complex(a, b)
    z = 0
    n = 0
    while abs(z) <= 2 and n < MAX_ITERATIONS:
        z = z*z + c
        n += 1
    pixel_color = 255 - int(n * 255 / MAX_ITERATIONS)
    return pixel_color

# dimensions of the heatmap
width = 600
height = 400

# range of values for the real and imaginary parts
real_range = np.linspace(-2, 1, width)
imaginary_range = np.linspace(-1, 1, height)

# array to store the colors of each pixel
heatmap = np.empty((height, width))

# color for each pixel
for i in range(width):
    for j in range(height):
        a = real_range[i]
        b = imaginary_range[j]
        heatmap[j, i] = mandelbrot(a, b)

# Plotting the heatmap
plt.figure(figsize=(10, 6))
plt.imshow(heatmap, cmap='hot', extent=[-2, 1, -1, 1])
plt.colorbar(label='Iterations')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Mandelbrot Fractal Heatmap')
plt.show()
```
